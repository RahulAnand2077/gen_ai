```python
from dotenv import load_dotenv
load_dotenv()

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
from langchain_core.messages import SystemMessage, HumanMessage

msg = [
    SystemMessage(content="You are a professsional guitarist and you have to help anyone with their queries about instrumental music."),
    HumanMessage(content="How one should practice to perfect hybrid picking?")
]

res = llm.invoke(msg)
print(res.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate

template = ChatPromptTemplate.from_messages([
    ("system", "You are a translator. Translate {input_language} to {output_language}."),
    ("human", "{text}")
])

print(f"Required variables: {template.input_variables}")
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
raw_msg = llm.invoke("Hi")
print(f"Raw Type: {type(raw_msg)}")

clean_text = parser.invoke(raw_msg)
print(f"Parsed Type: {type(clean_text)}")
print(f"Content: {clean_text}")
```

```python
from dotenv import load_dotenv
load_dotenv()

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser


llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
template = ChatPromptTemplate.from_template("Tell me a fun fact about {topic}.")
parser = StrOutputParser()
```

```python
prompt_value = template.invoke({"topic": "Djent Instrumental Music"})
response_obj = llm.invoke(prompt_value)
final_text = parser.invoke(response_obj)
print(final_text)
```

```python
chain = template | llm | parser
print(chain.invoke({"topic": "Economy & Hybrid picking"}))
```
