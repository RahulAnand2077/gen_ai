```python
import transformers
from transformers import pipeline
```

```python
plots = [
    "A disgraced special forces operative, now working as a high-stakes courier, accepts a job to transport a mysterious briefcase from Berlin to Paris. When he is ambushed by a team of mercenaries on the Autobahn, he realizes the package contains codes for a prototype weapon. With his car destroyed and enemies closing in from all sides, he must rely on his combat skills and a stolen motorcycle to outrun a corrupt government agency and deliver the evidence to the only reporter who can clear his name.",
    "Dave, a socially awkward accountant, accidentally crashes a high-profile celebrity wedding while trying to find a bathroom at a hotel. Mistaken for the eccentric wedding planner who went missing, Dave is forced to make split-second decisions about floral arrangements and seating charts to avoid being thrown out. As the ceremony descends into chaos involving a runaway llama and a chocolate fountain explosion, Dave tries to keep up the charade long enough to impress a bridesmaid, only to realize the real planner has just arrived.",
    "A group of geology students treks into a remote forest to investigate a newly discovered sinkhole. As they rappel down into the darkness, their equipment begins to malfunction, and they hear a rhythmic scratching sound echoing from the walls. One by one, the students vanish, only to reappear hours later, standing perfectly still in the shadows and mimicking the voices of their friends. Realizing they are not alone in the caves, the survivors must climb back to the surface before the thing in the deep steals their voices—and their lives—forever."
]
# labels = ["Action", "Comedy", "Horror"]
labels = ["Action movie", "Comedy movie", "Horror movie"]
```

```python
classifier = pipeline('zero-shot-classification',model='MoritzLaurer/DeBERTa-v3-base-mnli-fever-anli')
results = classifier(
    plots,
    labels,
    hypothesis_template="The genre of this movie is {}."    
)

for result in results:
    print(f"Plot Snippet: {result['sequence'][:50]}...")
    print(f"Top Prediction: {result['labels'][0]} ({result['scores'][0]:.2%})\n")
```

```python
classifier = pipeline('zero-shot-classification',model='distilbert-base-uncased')
results = classifier(
    plots,
    labels,
    hypothesis_template="The genre of this movie is {}."
)

for result in results:
    print(f"Plot Snippet: {result['sequence'][:50]}...")
    print(f"Top Prediction: {result['labels'][0]} ({result['scores'][0]:.2%})\n")
```
